```python
import numpy as np

# a = np.array([1,2,3,4])
# b = np.array([5,6,18,91])

a = np.array(template_1a)
b = np.array(template_1b)


def DTW(Ref_sig, Test_sig):
    n = len(Ref_sig)
    m = len(Test_sig)
    
    cost_matrix = np.zeros((n+1,m+1))
        
    for j in range(m+1):
        cost_matrix[0,j] = np.inf
    
    for i in range(n+1):
        cost_matrix[i,0] = np.inf

        
    cost_matrix[0,0] = 0
    
    distance = np.zeros((n,m))
    
    for i in range(0,n):
     
     for j in range(0,m):
            
            dist = np.linalg.norm(Test_sig[j]-Ref_sig[i])
            distance[i,j] = dist
            
    print(distance)
    path_matrix = np.zeros((n,m))
    for i in range(1,n+1):
        
        for j in range(1,m+1):
            min_dis = min(cost_matrix[i-1,j-1]  ,cost_matrix[i,j-1], cost_matrix[i-1,j]) 
            
            cost_matrix[i,j] = distance[i-1,j-1]+ min_dis
            
            if min_dis ==cost_matrix[i-1,j-1]:
                 path_matrix[i-1,j-1]=0 
            elif min_dis ==cost_matrix[i,j-1]:
                path_matrix[i-1,j-1]=1
            else:
                path_matrix[i-1,j-1]=2
                
            
            #match(i-1,j-1)
            #deletion(i,j-1)
            #insertion(i-1,j)
            
            
    i = n-1
    j = m-1
    path=[]
    while i >=0 or j>=0:
        path.append((i,j))
        check = path_matrix[i,j]
        if check ==0:
            i=i-1
            j=j-1
        elif check ==1:
            j=j-1
        else:
            i = i-1
    print(cost_matrix)
    print(path_matrix)
    print(path)
    

    
    return cost_matrix[n,m]
    
DTW(a,b)
            
    
```
